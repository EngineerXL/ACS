Построим графики для исходного и приведённого уравнения

```
fu[x1_, y1_, z1_] := u/.{x -> x1, y -> y1, z -> z1}

Print[fu[x, y, z], "= 0"]

RegionPlot3D[fu[x, y, z] <= 0, {x, -10, 10}, {y, -10, 10}, {z, -10, 10}]
```

```
fv[x1_, y1_, z1_] := v/.{x -> x1, y -> y1, z -> z1}

Print[fv[x, y, z] // N, " = 0"]

RegionPlot3D[fv[x, y, z] <= 0, {x, -10, 10}, {y, -10, 10}, {z, -10, 10}]
```
