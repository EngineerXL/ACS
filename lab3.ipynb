### Падение тела

```python
var("y1 y2 y3 y4 t")

tmax = 1.41
alpha = pi/4
v0 = 10.0
y0 = [0, 0, v0 * cos(alpha), 0, v0 * sin(alpha)]
k = 0.01
g = 9.81

ode = [y2, -k * y2 * sqrt(y2 ^ 2 + y4 ^ 2), y4, -k * y4 * sqrt(y2 ^ 2 + y4 ^ 2) - g]

solution = desolve_system_rk4(des = ode, vars = [y1, y2, y3, y4], ics = y0, ivar = t, step = 0.01, end_points = tmax)

lines = plot([])

for i in range(len(solution) - 1):
    lines += line([[solution[i][1], solution[i][3]], [solution[i + 1][1], solution[i + 1][3]]])

show(lines)

forget("y1 y2 y3 y4 t")
```

### Двухвидовая модель «хищник – жертва»

```python
var("x y")

b = [4, 3, 2]
lines = plot([])
colors = ["blue", "green", "red"]

for i in range(len(b)):
    eq1 = (3 - b[i] * y) * x
    eq2 = (-1 + 1 * x) * y
    solution = desolve_odeint(des = [eq1, eq2], ics = [2, 1], times = srange(0, 7, 0.01), dvars = [x, y])
    lines += line(solution, color = colors[i])
lines.show()

forget("x y")
```
